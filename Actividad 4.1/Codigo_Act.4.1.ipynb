```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Carga de archivo 
df = pd.read_csv('VentasTotales.csv')

#Verificamos la información del DataFrame
df.dtypes
```

```python
#Corroboramos valores nulos
df.isnull().sum().sum()
```

PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME

```python
#Creo 2 dataframe para poder procesar los outliers
cuantitativas= df.iloc[ : , 1:23 ]
cualitativas= df.iloc[ : , [0] ]
```

```python
#Verificar que esten las columnas en los dataframe asigandos 
cualitativas.columns
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=cuantitativas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
df_outliers= cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df_outliers
```

```python
#Corroboramos valores nulos
df_outliers.isnull().sum()
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data=df_outliers.copy()
data=df_outliers.fillna(round(df_outliers.mean(),1))
data
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
data.isnull().sum()
```

```python
#Unimos el dataframe cuantitativo limpio con el dataframe cualitativo
df_limpio = pd.concat([cualitativas, data], axis=1)
```

```python
#Convertir DataFrame a CSV
df_limpio.to_csv("VentasLimpias.csv")
```
